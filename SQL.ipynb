**Descobrindo a idade do cliente**

```python
#no MySQL esse código funcionaria.
#select date(now) - date(DATA_NASCIMENTO) as Idade from tbCliente;
#Mas no sqlite a sintaxe é a sequinte:
#A questão de ter um código limpo e indentado e que não coloque muitos códigos na mesma linha
#mas no ambiente colab usando o SQLITE quando pulo linhas ele da erro.
%sql SELECT STRFTIME('%Y', 'now') - STRFTIME('%Y', DATA_NASCIMENTO) - (STRFTIME('%m-%d', 'now') < STRFTIME('%m-%d', DATA_NASCIMENTO)) AS Idade FROM tbCliente;
```

```python
#Exibindo dados com condicional no caso só esta sendo mostrado
#os valores nos quais a idade do cliente é maior que 22 anos
%sql select * from tbcliente where idade >= 30;
```

**Limite de crédito de cliente entre 20000 e 25000**

Observe que 100% dos que possuem crédito entre 20000 e 25000 tem mais de 30 anos de idade.

```python
%sql select * from tbCliente where LIMITE_CREDITO between 20000 and 25000;
```

```python
%sql select * from tbcliente where DATA_NASCIMENTO > '1995-01-13';
```

```python
#No MySQL a sintaxe funcionária somente passando YEAR mas no SQLITE funciona com STRTIME
#Isso irá formatar a data de nascimento no formato de ano (%Y) e então podemos comparar com o ano desejado, neste caso, '1995'.
#%sql select * from tbcliente where year(DATA_NASCIMENTO) = 1996;
%sql SELECT * FROM tbcliente WHERE strftime('%Y', DATA_NASCIMENTO) > '1995';
```

Filtragem por BAIRRO

```python
%sql select * from tbCliente where BAIRRO in ('Boa Viagem','Boa Vista','Caratinga');
```

Filtragem de idades maior ou iguais a 18 e menor ou igual a 22, sexo masculino ou morador da cidade Barra da tijuca ou Caratinga

```python
%sql SELECT * FROM tbCliente WHERE (IDADE >= 18 AND IDADE <= 22 AND SEXO = 'M') OR (CIDADE = 'Barra da Tijuca' or CIDADE = 'Caratinga');
```

Consulta de limite de crédito por Estado

```python
%sql select ESTADO, sum(LIMITE_CREDITO) as soma_limite from tbCliente group by estado having sum(LIMITE_CREDITO) order by estado desc;
```

Volume de crédito por cidade

```python
%sql Select Cidade, SUM(LIMITE_CREDITO) as limite_cidade From tbCliente group by cidade order by limite_cidade desc;
```

**Podemos observar que a ordem de limite de crédito é bem parecida com a ordem de volume total de vendas, indicando que quanto mais crédito o cliente tem mais ele compra.**

```python
#Volume de vendas por cidade
%sql select CIDADE, sum(VOLUME_COMPRA) as volume_total from tbCliente group by CIDADE order by volume_total desc;
```

**Observe que 73% do nosso volume de venda são de clientes com idade acima de 29 anos**

```python

%sql WITH total_vendas AS (SELECT SUM(VOLUME_COMPRA) AS total FROM tbCliente), vendas_maior_29 AS (SELECT SUM(VOLUME_COMPRA) AS total FROM tbCliente WHERE IDADE > 29) SELECT (vendas_maior_29.total * 100.0 / total_vendas.total) AS percentual FROM total_vendas, vendas_maior_29;
```

**12% do nosso limite de crédito pertence a pessoas com menos de 29 anos de idade**

```python
%sql WITH total_credito AS (SELECT SUM(limite_credito) AS total FROM tbCliente),limite_idade AS (SELECT SUM(limite_credito) AS total FROM tbCliente WHERE idade < 29) SELECT (limite_idade.total * 100.0 / total_credito.total) AS percentual FROM total_credito, limite_idade;
```

**Se subissemos o limite de crédito dos clientes menores que 29 anos teriamos um lucro maior**
